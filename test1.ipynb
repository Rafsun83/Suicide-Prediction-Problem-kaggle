```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as splot
import matplotlib.pyplot as plt
import seaborn
```

```python
df=pd.read_csv('foreveralone.csv')
```

```python
df.head()
```

```python
df['attempt_suicide'].value_counts()
```

```python
df.shape
```

```python
x=df.iloc[:,:-5]
y=df.attempt_suicide
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
x.loc[:,['time','gender','sexuallity','race','bodyweight','virgin','prostitution_legal','pay_for_sex','social_fear','depressed','what_help_from_others','friends','income']]=\
x.loc[:,['time','gender','sexuallity','race','bodyweight','virgin','prostitution_legal','pay_for_sex','social_fear','depressed','what_help_from_others','friends','income']].apply(enc.fit_transform)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=20)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
pd.crosstab(y_test,y_predict)
```
